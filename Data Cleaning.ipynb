## Data Cleaning Workflow

### 1. Import Required Libraries

- Display the first few records.

### 10. Final Data Quality Check

- Verify duplicate removal.

**Data Cleaning & Preprocessing using Python (Pandas)**

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("Raw Dataset for Data Analytics.xlsx")
```

```python
df.head()
df.shape
df.info()
df.describe(include='all')
```

```python
df.isnull().sum()

missing = (df.isnull().sum()/len(df))*100
missing.sort_values(ascending=False)
```

```python
df.duplicated().sum()
df = df.drop_duplicates()
```

```python
df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
```

```python
text_cols = df.select_dtypes(include='object').columns

for col in text_cols:
    df[col] = df[col].str.title()
```

```python
df['CouponCode'] = df['CouponCode'].fillna("No Coupon")
```

```python
df.dtypes
```

```python
numeric_cols = df.select_dtypes(include=['int64','float64']).columns

numeric_cols
```

```python
df[df['Quantity'] < 0]
df[df['UnitPrice'] < 0]
```

```python
df['Expected_Total'] = df['Quantity'] * df['UnitPrice']

wrong_total = df[df['Expected_Total'] != df['TotalPrice']]

wrong_total
```

```python
Q1 = df['UnitPrice'].quantile(0.25)
Q3 = df['UnitPrice'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df['UnitPrice'] < lower) | (df['UnitPrice'] > upper)]

outliers
```

```python
for col in text_cols:
    print(col)
    print(df[col].unique())
    print("-"*40)
```

```python
df.columns
df.columns = df.columns.str.strip()
df.columns = df.columns.str.replace(" ", "_")
```

```python
df.info()
df.describe()
df.isnull().sum()
```

```python
df.to_excel("Cleaned_Dataset.xlsx", index=False)
```
